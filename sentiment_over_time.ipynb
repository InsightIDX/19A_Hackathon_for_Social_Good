```python
import pandas as pd
import bs4 as bs 
import numpy as np
import os
from PIL import Image
import codecs, json
import pandas as pd
```

```python
curr_dir = os.getcwd()
tar_dir = curr_dir + '/data/sentiment'
os.listdir(tar_dir)
```

```python
file = os.listdir(tar_dir)[7]
path = tar_dir + '/' + file
file
```

```python
df = pd.read_csv(path, sep=',',error_bad_lines=False)
df.head()
```

```python
# change date to datetime object
df['date'] = pd.to_datetime(df.date)
```

```python
print(df.info())
df.head()
```

```python
# set datetime object to index
df = df.set_index('date')
```

```python
avg_sent_per_week = df.sentiment.resample('h').mean()
```

```python
avg_sent_per_week.plot()
```
